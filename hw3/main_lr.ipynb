```python
from base.gnb import GaussianNaiveBayes
from base.lr import LogisticRegression
from base.utils import confusion_matrix, sens, spec, acc
from data import data
import numpy as np
%matplotlib inline
import matplotlib

matplotlib.interactive(False)
from matplotlib import pyplot


if __name__ == '__main__':
    train, test = data.load_news()
    train = train['train']
    test = test['test']
    features = train.shape[1] - 1
    list_lambda = [2, 1, 0, -1, -2, -3, -4, -5]
    list_errors = LogisticRegression.cross_validation_for_lambda(list_lambda, train[:, :-1], train[:, -1])
    lam = list_lambda[list_errors.index(min(list_errors))]
    print "Lambda|Error", np.c_[list_lambda, list_errors]
    print "Picking best lambda -->", lam
    model = LogisticRegression()
    model.fit(train[:, :-1], train[:, -1], num_iterations=10000, learning_rate=0.01, lam=10. ** lam, print_cost=True)

    # test accuracy
    y_true = test[:, -1]
    y_pred = model.predict(test[:, :-1])
    cm = confusion_matrix(y_true, y_pred)
    print "Test confusion matrix:"
    print cm
    print "Test error:"
    print 1 - acc(cm)

    # train accuracy
    y_true = train[:, -1]
    y_pred = model.predict(train[:, :-1])
    cm = confusion_matrix(y_true, y_pred)
    print "Train confusion matrix:"
    print cm
    print "Train error:"
    print 1 - acc(cm)

    # draw roc
    y_true = test[:, -1]
    y_score = model.predict(test[:, :-1], score=True)
    score_min = min(y_score)
    score_max = max(y_score)
    area = score_max - score_min
    roc_data = []
    for score in np.arange(score_min, score_max, area / 20.0):
        y_pred = map(lambda x: 1. if x > score else -1., y_score)
        cm = confusion_matrix(y_true, y_pred)
        roc_data.append([1 - sens(cm), spec(cm)])
    roc_data = np.asarray(roc_data)
    pyplot.plot(roc_data[:, 0], roc_data[:, 1])
    pyplot.axes().set_aspect('equal')
    pyplot.xlim([0,1])
    pyplot.ylim([0,1])
    pyplot.show()
```
